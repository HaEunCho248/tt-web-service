```python
fig,ax=plt.subplots(3,3)
fig.set_size_inches((20, 15))
fig.suptitle('환경오염 방지 노력', fontsize = 15)
plt.subplots_adjust(wspace = 0.3, hspace = 0.3)
ax[0,0].plot(effort.index,effort["합성세제 사용 줄이기"],label="합성세제 사용 줄이기")
ax[0, 0].set_title('합성세제 사용 줄이기', fontsize = 12)

ax[0,1].plot(effort.index,effort["음식물 쓰레기 줄이기"],label="음식물 쓰레기 줄이기")
ax[0, 1].set_title('음식물 쓰레기 줄이기', fontsize = 12)

ax[0,2].plot(effort.index,effort["대중교통을 이용하기"],label="대중교통을 이용하기")
ax[0, 2].set_title('대중교통을 이용하기', fontsize = 12)

ax[1,0].plot(effort.index,effort["재활용품 분리하여 배출하기"],label="재활용품 분리하여 배출하기")
ax[1, 0].set_title('재활용품 분리하여 배출하기', fontsize = 12)

ax[1,1].plot(effort.index,effort["일회용품 사용 줄이기"],label="일회용품 사용 줄이기")
ax[1,1].set_title('일회용품 사용 줄이기', fontsize = 12)

ax[1,2].plot(effort.index,effort["친환경제품 구입·사용하기"],label="친환경제품 구입·사용하기")
ax[1,2].set_title('친환경제품 구입·사용하기', fontsize = 12)

ax[2,0].plot(effort.index,effort["자연보호 및 환경보전활동 참여하기"],label="자연보호 및 환경보전활동 참여하기")
ax[2,0].set_title('자연보호 및 환경보전활동 참여하기', fontsize = 12)

ax[2,1].plot(effort.index,effort["물 절약하기"],label="물 절약하기")
ax[2,1].set_title('물 절약하기', fontsize = 12)

ax[2,2].plot(effort.index,effort["가정내 대기전력 줄이기"],label="가정내 대기전력 줄이기")
ax[2,2].set_title('가정내 대기전력 줄이기', fontsize = 12)

```

```python
greenGas["온실가스 총합"]=greenGas["간접배출 온실가스"]+greenGas["직접배출 온실가스"]
greenGas["에너지환산량 총합"]=greenGas["간접배출 에너지환산량"]+greenGas["직접배출 에너지환산량"]

greenGas
```

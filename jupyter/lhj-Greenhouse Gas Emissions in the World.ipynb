```python
ggkor = gg.iloc[[14]].drop(columns = ['Country', 'Data source', 'Sector', 'Gas', 'Unit'])

ggworld = gg.iloc[[0]].drop(columns = ['Country', 'Data source', 'Sector', 'Gas', 'Unit'])

# ggkor = ggkor.sort_values(by=ggkor.columns, ascending=True)
display(ggkor)
```
